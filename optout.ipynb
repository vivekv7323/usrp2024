```python
from scipy.optimize import curve_fit
import pandas as pd
import matplotlib.pyplot as plt


t = pd.read_csv("data.txt", sep="\t")
t = t.sort_values(by="x", axis=0)
def quadratic(x, a, b, c):
    return a+b*x+c*x*x

fit = curve_fit(quadratic, t["x"], t["y"], sigma=t["y_std"])

plt.scatter( t["x"], t["y"], s=1, label="data")
plt.plot(t["x"], quadratic(t["x"], fit[0][0], fit[0][1], fit[0][2]), label="fit")
plt.legend(loc="best")
plt.xlabel("x")
plt.ylabel("y")
plt.show()
```
